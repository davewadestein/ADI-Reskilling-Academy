```python
def rounder25(amount):
    """Return amount rounded UP to nearest
    quarter dollar.
       
    ...$1.89 becomes $2.00
    ...but $1.00/$1.25/$1.75/etc.
    remain unchanged
    """
    dollars = int(amount) # 1
    cents = round((amount - dollars) * 100) # 8
    
    quarters = cents // 25 # 3
    if cents % 25: # 14
        quarters += 1 # 4
        
    amount = dollars + 0.25 * quarters # 2.00

    return amount
```

* Write a function which takes a number as a parameter and returns a string version of the number with commas representing thousands, e.g., __`add_commas(12345)`__ would return "12,345"
* Write a function to demonstrate the Collatz Conjecture:
  * for integer n > 1
    * if n is even, then __`n = n // 2`__
    * if n is odd, then __`n = n * 3 + 1`__
  * ...will always converge to 1
  * (your function should take n and keep printing new value of n until n is 1)


```python
def sum_digits(num):
    """Sum up digits and if result is more than 1 digit, keep summing until result is 1 digit."""
    # debugging
    print(f'sum_digits({num})')
    # Much easier (for me) to treat the number as a string and then pick off each 'digit'
    result = 0
    
    for digit in str(num): # convert to num so we can iterate through the number
        result += int(digit) # convert back to int to do the addition
    
    if result > 9: # i.e., > 1 digit
        return sum_digits(result) # call ourselves "recursively" to get the job done
    
    return result # 1 digit, so just return it
```

```python
def add_commas(num):
    """Return a string representing the number with commas added for thousands."""
    num = str(num)[::-1] # string-ify and reverse to make it easier to work with
    comma_num = '' # result will be built up in this string
    
    for position in range(0, len(num), 3): # go through string 3 digits at a time
        comma_num += num[position:position + 3] # append the current 3 digits
        if position < len(num) - 3: # if we're not at the end, add a comma
            comma_num += ','
    
    return comma_num[::-1] # reverse back 
```

```python
add_commas(1123712345)
```

```python
def collatz(num):
    """Show sequence of Collatz Conjecture...blah"""
    if type(num) == str:
        num = int(num)
    count = 0 # for formatting, we will limit to 10 numbers per line of output
    while num != 1:
        print(f'{num:9d}', end='')
        count += 1
        if count % 10 == 0: # every 10 numbers, print a newline
            print()
        if num % 2 == 0: # fails if num is float
            num //= 2 
        else:
            num = num * 3 + 1
    print('        1')
```

```python
def product(*terms): # 0+ terms
    """multiply all terms together and return result"""
    result = 1
    
    for term in terms: # iterate through the terms (args) tuple
        result *= term
        
    return result
```

```python
def sum_digits(num, **options):
    """Sum up digits and if result is more than 1 digit, keep summing until result is 1 digit.
    
    Keyword args (options):
        debug = level of debugging ('none', 'some', 'all')
    """
    debug_level = 'none'
    
    if 'debug' in options: # user passed this in, i.e., debug=...
        if options['debug'] in { 'none', 'some', 'all' }:
            debug_level = options['debug'] # grab what the user passed in
        else:
            print('unknown debugging level:', options['debug'])
    
    if debug_level in { 'some', 'all' }:
        print(f'DEBUG: sum_digits({num}, options={options})')
                  
    # Much easier (for me) to treat the number as a string and then pick off each 'digit'
    result = 0
    
    for digit in str(num): # convert to num so we can iterate through the number
        result += int(digit) # convert back to int to do the addition
    
    if debug_level == 'all':
        print('DEBUG: Intermediate sum =', result)
    
    if result > 9: # i.e., > 1 digit
        if debug_level == 'all':
            print(f'DEBUG: calling sum_digits({result}, debug={debug_level})')
        return sum_digits(result, debug=debug_level) # call ourselves "recursively" to get the job done
    
    return result # 1 digit, so just return it
```

## Exceptions: __`try/except`__
* __`try`__ block wraps code which may throw an exception, and __`except`__ block catches exception

```python
try:
    mylist[5] # could throw an IndexError
except:
    print('no element at offset 5')
    # cleanup, reset, ...
    
print('rest of program')
```

```python
mylist[5]
```

* problem? above example catches ALL exceptions, not just __`IndexError`__ we are expecting
* best practice is to catch expected exceptions and let unexpected ones through, so as to avoid hidden errors

```python
try:
    print(mylist[5]) # could throw an IndexError
    int('a') # wrongly believe this can't throw an exception
except IndexError:
    print('Bad index! Try again!')
except Exception as uhoh: # put the exception into the variable uhoh
    print('Some other exception:', uhoh, type(uhoh))
```

```python
short_list = ['zero', 'one', 'two']

while True:
    value = input('Enter numeric index [q to quit]? ')
    if value == 'q':
        break
    try:
        position = int(value) # they could enter a non-int
        print(short_list[position]) # fall off the list... 
    except IndexError:
        print('Bad index:', value)
    except ValueError:
        print(f'You must enter an integer–{value} is not an integer!')
    except Exception as other:
        print('Something else broke:', other, type(other))
```

## Exceptions (cont'd)
* important to minimize size of try block


```python
# pseudocode
try:
    dangerous_call() # presumably could throw an exception
    after_call() # this can't throw an exception
except OS_Error:
    log('...')

# after_call() will only run if dangerous_call() doesn't throw
# an exception…So what's the problem?
```

```python
# pseudocode
try:
    dangerous_call()
except OS_Error:
    log('...')
else: # i.e., no exception
    after_call()
    
# now it’s clear that try block is guarding against possible errors in dangerous_call(), not  in after_call()
# it’s also more obvious that after_call() will only execute if no exceptions are raised in the try block
```

## __The `finally` Block__
* code in the finally block will be executed whether or not an exception is thrown

```python
def func():
    try:
        i = int(input('\nEnter a number: ')) # ValueError?
        x = 1 / i # ZeroDivisionError?
    except ValueError:
        print('Not a number!')
        return
    except ZeroDivisionError:
        print('Cannot divide by 0')
    else:
        print('Everything OK')
    finally:
        print('FINALLY: DO this either way!')

func(), func(), func()
```
